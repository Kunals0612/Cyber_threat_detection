```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame and it has a 'Date' column in 'YYYY-MM-DD' format
# Replace 'your_data.csv' with the path to your dataset file
# df = pd.read_csv('your_data.csv')

# Convert the 'Date' column to datetime and extract the year
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Filter for the 'Exploitation of Application Server' Sub-Type
exploit_df = df[df['Industry'] == 'Educational Services']

# Group by 'Year' and count the occurrences
yearly_counts = exploit_df.groupby('Year').size()

# The ARIMA model requires a univariate time series, so we'll use 'yearly_counts'
# The order (p,d,q) needs to be determined based on the data
# For this example, we'll use (1,1,1) which is a common starting point
model = ARIMA(yearly_counts, order=(1,1,1))

# Fit the model
model_fit = model.fit()

# Make forecast for the next 5 years
forecast = model_fit.forecast(steps=5)

# Convert the index to integer if it's not already
yearly_counts.index = yearly_counts.index.astype(int)

# Now create the range for the x-axis of the forecast plot
forecast_years = range(yearly_counts.index[-1] + 1, yearly_counts.index[-1] + 6)

# Plot the historical data
plt.figure(figsize=(12, 6))
plt.plot(yearly_counts.index, yearly_counts, label='Historical')

# Plot the forecast
plt.plot(forecast_years, forecast, label='Forecast', color='red')

# Add labels, title, and legend
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('ARIMA Forecast of External Denial of Service')
plt.legend()
plt.show()

model_fit = model.fit()

# Make forecast for the next 5 years
forecast = model_fit.forecast(steps=5)

# Convert the index to integer if it's not already
yearly_counts.index = yearly_counts.index.astype(int)

# Create the range for the forecasted years
forecast_years = range(yearly_counts.index[-1] + 1, yearly_counts.index[-1] + 6)

# Print the forecasted values
for year, value in zip(forecast_years, forecast):
    print(f"Year: {year}, Forecasted Count: {value}")
```
